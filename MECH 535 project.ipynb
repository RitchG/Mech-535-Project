```python
import numpy as np
import math
from matplotlib import pyplot as plt

incompressible = 0

#defining the given parameters 
rotor_width = 0.1  #in m
N_rpm = 6000  #rpm
N_rad_s = (N_rpm * 2 * np.pi) / 60  #converting to rad/s
Cx = 136  #in m/s
density_inlet = 1.5  #kg/m^3 (static density inlet)
temperature_inlet = 25 + 273.15  #in kelvins 
blade_tip_diameter = 1.0  #in m
blade_hub_diameter = 0.9  #in m
alpha_flow_inlet_degrees = np.degrees(30)  #flow inlet angle in degrees
alpha_flow_inlet_rad = np.radians(30)  #flow inlet angle in radians
swirl_increase_hub = 82.3  #m^2/s at hub
swirl_increase_tip = 84.4  #m^2/s at tip
loss = 0.05  #5% loss factor

#constants that we are going to be using 
R = 287  #in J/(kg*K) for air
gamma = 1.4  #isentropic exponent
cp = 1005 # J/(kg*K) for air

#finding the radius of the hub and the tip
RHUB = blade_hub_diameter / 2
RSHROUD = blade_tip_diameter / 2

N = 0 + 5*10 + 10 + 5*10
del_x = rotor_width/10
del_r = del_x
HUBTOSHROUD = RSHROUD - RHUB 
M = round(HUBTOSHROUD/del_r)

x = np.linspace(0-50, N-50, N+1)/100
r = np.linspace(RHUB, RSHROUD, M+1)

def PSI(r):
    return (r**2-RHUB**2)/(RSHROUD**2-RHUB**2)

psi = np.ones(((M+1, N+1)))
for j in range(M+1):
    psi[j] = PSI(r[j])

DENSITY_INLET = 1.5
CM_INLET = 136
Cp = 1.0005
T0_INLET = 25 #not sure
ALPHA = 30 # degrees
C_INLET = CM_INLET/np.cos((ALPHA*np.pi)/180) #is CM = CX???

A_INLET = 2*np.pi*(RSHROUD**2-RHUB**2)
XMASS = DENSITY_INLET*CM_INLET*A_INLET
H0_INLET = Cp*temperature_inlet + C_INLET**2/2

def cx_calc(rho, rij, psi_up, psi_down):
    cx = (XMASS/(2*np.pi*rho*rij))*((psi_up-psi_down)/(2*0.01))
    return cx

def cr_calc(rho, rij, psi_right, psi_left):
    cx = (XMASS/(2*np.pi*rho*rij))*((psi_right-psi_left)/(2*0.01))
    return cx

CX = Cx*np.ones((M+1, N+1))
CR = np.zeros((M+1, N+1))
CM = CX

rCthetha = np.ones(((M+1, N+1)))
for j in range(M+1):
    rCthetha[j] = r[j]*Cx*np.tan((ALPHA*np.pi)/180)

OMEGLOS_ROTOR = np.ones((6, 10))*(0.5/(10))

OMEGLOS_BEFORE = np.zeros((6, 51))
OMEGLOS_AFTER = np.zeros((6, 50))
OMEGLOS = np.concatenate((np.concatenate((OMEGLOS_BEFORE,OMEGLOS_ROTOR),axis=1),OMEGLOS_AFTER),axis=1)

del_rCthethaTE = np.zeros(M+1)
for j in range(M+1):
    del_rCthethaTE[j] = 82.3+(84.4-82.3)*(j)/(M+1-1)
    
iLE = 50
iTE = iLE+10

for i in range(iLE+1, iTE+1):
    for j in range(M+1):
        rCthetha[j, i] = rCthetha[j,i-1] + del_rCthethaTE[j]*(i-iLE)/(iTE-iLE)

T = T0_INLET*np.ones((6, N+1))
S = np.zeros((6, N+1))
H0 = H0_INLET*np.ones((6, N+1))
rho = DENSITY_INLET*np.ones((6, N+1))

# start iterations here
iteration = 0
while iteration < 200:
    iteration += 1
    psi_old = psi
    rho_old = rho
    VORTICITY_old = VORTICITY
    VORTICITY = np.zeros((M+1, N+1))
    for j in range(1, M):
        for i in range(51, N+1):
            VORTICITY[j,i] = (np.pi/(del_r*XMASS*CX[j,i]))*((rCthetha[j,i]/r[j]**2)*(rCthetha[j+1,i]-rCthetha[j-1,i])+T[j, i]*(S[j+1, i]-S[j-1, i])-(H0[j+1, i]-H0[j-1, i])) 

    S = np.zeros((6, N+1))
    A = np.zeros((4, N+1))
    B = np.zeros((4, N+1))

    if incompressible:
        for j in range(1, M):
            for i in range(1, N):
                A[j-1,i] = DENSITY_INLET/((1/r[j])+(1/r[j])+(1/(r[j]+0.005))+(1/(r[j]-0.005)))
                B[j-1,i] = (1/DENSITY_INLET)*((psi[j,i+1]/r[j])+(psi[j,i-1]/r[j])+(psi[j+1,i]/(r[j]+0.005))+(psi[j-1,i]/(r[j]-0.005)))
                psi[j,i] = A[j-1,i]*(B[j-1,i] + 0.01**2*VORTICITY[j,i])
    else:
        for j in range(1, M):
            for i in range(1,N):
                A[j-1,i] = 1/((1/(((rho[j,i]+rho[j,i+1])/2)*r[j]))+(1/(((rho[j,i]+rho[j,i-1])/2)*r[j]))+(1/(((rho[j,i]+rho[j+1,i])/2)*(r[j]+0.005)))+(1/(((rho[j-1,i]+rho[j,i])/2)*(r[j]-0.005))))
                B[j-1,i] = ((psi[j,i+1]/(((rho[j,i]+rho[j,i+1])/2)*r[j]))+(psi[j,i-1]/(((rho[j,i]+rho[j,i-1])/2)*r[j]))+(psi[j+1,i]/(((rho[j,i]+rho[j+1,i])/2)*(r[j]+0.005)))+(psi[j-1,i]/(((rho[j-1,i]+rho[j,i])/2)*(r[j]-0.005))))
                psi[j,i] = A[j-1,i]*(B[j-1,i] + 0.01**2*VORTICITY[j,i]) 

    for j in range(M+1):
        for i in range(1,N):
            if j == 0:
                CX[j,i] = (XMASS/(2*np.pi*rho[j,i]*r[j]))*((psi[j+1,i]-psi[j,i])/(2*0.01))
                CR[j,i] = 0
                CM[j,i] = (CR[j,i]**2+CX[j,i]**2)**0.5
            if j == 5:
                CX[j,i] = (XMASS/(2*np.pi*rho[j,i]*r[j]))*((psi[j,i]-psi[j-1,i])/(2*0.01))
                CR[j,i] = 0
                CM[j,i] = (CR[j,i]**2+CX[j,i]**2)**0.5
            else:  
                CX[j,i] = (XMASS/(2*np.pi*rho[j,i]*r[j]))*((psi[j+1,i]-psi[j-1,i])/(2*0.01))
                CR[j,i] = (-XMASS/(2*np.pi*rho[j,i]*r[j]))*((psi[j,i+1]-psi[j,i-1])/(2*0.01))
                CM[j,i] = (CR[j,i]**2+CX[j,i]**2)**0.5

    h0rel = np.zeros((6, N+1))
    Ctheta = np.zeros((6, N+1))
    Vtheta = np.zeros((6, N+1))
    beta = np.zeros((6, N+1))
    Vsquared = np.zeros((6, N+1))
    h = np.zeros((6, N+1))
    P_oreli = np.zeros((6, N+1))
    P_orel = np.zeros((6, N+1))
    P0 = np.zeros((6, N+1))
    P = np.zeros((6, N+1))

    for j in reversed(range(M+1)):
        for i in reversed(range(51,N+1)):
            a = (psi[j,i]-psi[j-1,i-1])/(psi[j,i-1]-psi[j-1,i-1])
            b = 1 - a
            if i >= 51 and i <= 61 and i-1 >= 51 and i-1 >= 61: # if inside rotor
                if j == 0:
                    h01rel = a*h0rel[j,i-1]
                    h0rel[j,i] = h01rel - ((N_rad_s*a*r[j])**2)/2 + ((N_rad_s*r[j])**2)/2
                    P01rel = a*P_orel[j,i-1]
                    P1 = a*P[j,i-1]
                    s1 = a*S[j,i-1]
                else:
                    h01rel = a*h0rel[j,i-1]+b*h0rel[j-1,i-1]
                    h0rel[j,i] = h01rel - ((N_rad_s*(a*r[j]+b*r[j-1]))**2)/2 + ((N_rad_s*r[j])**2)/2
                    P01rel = a*P_orel[j,i-1]+b*P_orel[j-1,i-1]
                    P1 = a*P[j,i-1]+b*P[j-1,i-1]
                    s1 = a*S[j,i-1]+b*S[j-1,i-1]
                Ctheta[j,i] = rCthetha[j,i]/r[j]
                Vtheta[j,i] = N_rad_s*r[j] - Ctheta[j,i]
                beta[j,i] = math.atan(Vtheta[j,i]/CM[j,i])
                Vsquared[j,i] = CM[j,i]**2 + Vtheta[j,i]**2
                h[j,i] = h0rel[j,i] - Vsquared[j,i]/2
                H0[j,i] = h[j,i] + (CM[j,i]**2+Ctheta[j,i]**2)/2
                P_oreli[j,i] = P01rel*(h0rel[j,i]/h01rel)**(gamma/(gamma-1))
                P_orel[j,i] = P_oreli[j,i] - OMEGLOS[j,i]*(P01rel-P1)
                P0[j,i] = P_orel[j,i]*(H0[j,i]/h0rel[j,i])**(gamma/(gamma-1))
                P[j,i] = P0[j,i]*(h[j,i]/H0[j,i])**(gamma/(gamma-1))
                rho[j,i] = (cp*P[j,i])/(R*h[j,i])
                S[j,i] = s1 + cp*math.log(h0rel[j,i]/h01rel) - R*math.log(P_orel[j,i]/P01rel)
                T[j,i] = P[j,i]/(rho[j,i]*R)
            else: # if outside rotor
                if j == 0:
                    h01 = H0[j,i-1]
                    s1 = a*S[j,i-1]
                    P01 = a*P0[j,i-1]
                else:
                    h01 = a*H0[j,i-1]+b*H0[j-1,i-1]
                    s1 = a*S[j,i-1]+b*S[j-1,i-1]
                    P01 = a*P0[j,i-1]+b*P0[j-1,i-1]
                Ctheta[j,i] = rCthetha[j,i]/r[j]
                Vtheta[j,i] = - Ctheta[j,i]
                beta[j,i] = math.atan(Vtheta[j,i]/CM[j,i])
                Vsquared[j,i] = CM[j,i]**2 + Vtheta[j,i]**2
                h[j,i] = H0[j,i] - Vsquared[j,i]/2
                H0[j,i] = h[j,i] + (CM[j,i]**2+Ctheta[j,i]**2)/2
                P[j,i] = P0[j,i]*(h[j,i]/H0[j,i])**(gamma/(gamma-1))
                print(H0[j,i])
                rho[j,i] = (cp*P[j,i])/(R*h[j,i])
                S[j,i] = s1 + cp*math.log(H0[j,i]/h01) - R*math.log(P0[j,i]/P01)
                T[j,i] = P[j,i]/(rho[j,i]*R)
            
    if iteration >= 50:
        if np.max(np.abs(psi - psi_old)) <= 10**(-5) or np.max(np.abs(rho - rho_old)) <= 10**(-3) or np.max(np.abs(VORTICITY - VORTICITY_old)) <= 10**(-3):
            break
            

[x_plot, r_plot] = np.meshgrid(x, r)

fig, ax = plt.subplots(1, 1)

# plots filled contour plot 
ax.contourf(x_plot, r_plot, psi) 
  
ax.set_title('Filled Contour Plot') 
ax.set_xlabel('x') 
ax.set_ylabel('r') 
  
plt.show()
print(r)
print(psi)
print(rho)
```
